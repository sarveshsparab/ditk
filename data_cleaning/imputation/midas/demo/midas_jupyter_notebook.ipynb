1) Import Packages

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import csv
from sklearn.metrics import mean_squared_error as mse
from math import sqrt
```

2) Import Midas

```python
from midas import Midas
```

3) Create instance of Midas class

```python
imputer = Midas(layer_structure= [128], vae_layer= False, seed= 42)
```

4) Original gold dataset

```python
data_old = pd.read_csv("letter-recognition.csv",header = None)
```

5) Introduce missingness in the data at 20% rate. Returns a dataframe.

```python
data=imputer.preprocess("letter-recognition.csv")
```

6) Scale down to [0,1] and fill Null values with Nan

```python
  scaler = MinMaxScaler()
  na_loc = data.isnull()
  data.fillna(data.median(), inplace= True)
  data = pd.DataFrame(scaler.fit_transform(data), columns= data.columns)
  data[na_loc] = np.nan
  columns_list = []
```

7) Build Model -
   This method is called to construct the neural network that is the heart of
   MIDAS. This includes the assignment of loss functions to the appropriate
   data types.

```python
  imputer.build_model(data, softmax_columns= columns_list)
```

8) Overimpute
   This function spikes in additional missingness, so that known values can be
   used to help adjust the complexity of the model.

```python
  imputer.overimpute(training_epochs= 100, report_ival= 1,report_samples= 5, plot_all= False)
```

9) Train -    
    This is the standard method for optimising the model's parameters. Must be
    called before imputation can be performed.

```python
imputer.train(train_data = imputer.imputation_target.values,training_epochs= 100, verbosity_ival= 1)
```

10) Call the impute function with parameter m = 5, that means create 5 sets of imputations

```python
    imputer.impute(m= 5)
```

11) Convert data from [0,1] back to original values, with removed values as NaN

```python
    data = pd.DataFrame(scaler.inverse_transform(data))
```

```python
    count=0
    rmse=0
```

12) Update values which are NaN in the dataframe to the mean of the imputed values. Also, calculate rmse between the original gold dataset and the one with imputed values.

```python
  for a in range(na_loc.shape[0]):
    for b in range (na_loc.shape[1]):
      if na_loc.at[a,b] == True:
          count+=1
          imputed_vals = []
          for dataset in imputer.output_list:  
            imputed_vals.append(pd.DataFrame(scaler.inverse_transform(dataset),
                                     columns= dataset.columns).loc[a][b])
          val=np.mean(imputed_vals)
          val=round(val,2)
          data.at[a,b] = val
          rmse+= (data.at[a,b] - data_old.at[a,b])**2
```

13) Calculate rmse

```python
  rmse=rmse/count
  rmse = sqrt(rmse)
  print(rmse)
```

14) Output to result.csv

```python
  data.to_csv("result.csv", index=False,header=False)
```
