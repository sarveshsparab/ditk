# Similarity Text Encoding

- Simple description of Similarity Text Encoding and this notebook

## 1. Import the nessecery libarary and files

```python
import os
import sys
import numpy as np

from text_embedding_similarity import SimilarityEncoding
```

## 2. Set up usage variables

```python
embedding_dimension = 300
te = SimilarityEncoding(embedding_dimension)
```

## 3. Read the data

```python
train_input = '../data/sample_data/input.txt'

te.read_Dataset(train_input)
```

## 4. Train on training data

```python
# train on training data
output = te.train()

print ("Training embedding output shape: ")
print (output.shape)
```

```python
# output training data to a file(train_output)
unWords = np.unique(te.input_data)
index = 0

train_output = '../data/sample_data/train_output.txt'
file = open(train_output,"w")
for x in output.tolist():
    file.write(str(unWords[index]) + ' ' )
    file.write(str(x))
    file.write("\n")
    index += 1
file.close()
```

```python
print ("Example output: ")
print(unWords[0])
print(output.tolist()[0])
```

## 5. Making prediction(embedding) using training data

```python
# Predict embeddings
predict_input = '../data/sample_data/predict.txt'
output = te.predict_embedding(predict_input)

print ("Predict embedding output shape: ")
print (output.shape)
```

```python
# output predicted embeddings to a file(train_output)
unWords = np.unique(te.predict_data)
index = 0

predict_output = '../data/sample_data/predict_output.txt'
file = open(predict_output,"w")
for x in output:
    file.write(str(unWords[index]) + ' ' )
    file.write(str(x))
    file.write("\n")
    index += 1
file.close()
```

```python
print ("Example output: ")
print(unWords[0])
print(output.tolist()[0])
```

## 6. Predict similarity between two strings

```python
# predict similarity between two strings
# similarity_input1, similarity_input2
similarity_input1 = 'Viterbi'
similarity_input2 = 'Engineering'

sim = te.predict_similarity(similarity_input1, similarity_input2)
print("similarity between \'" + similarity_input1 + "\' and \'" + similarity_input2 + "\' is: ")
print(sim)
```

## 7.Evaluation example using Semi2017 bechmark

```python
te.read_Dataset('../data/Semi2017/input.txt')
te.train()
result = te.evaluate("Semi2017",'../data/Semi2017/sts-test-2017-small.csv')
print("Mean Error of similarity prediction: ")
print(result)
```
